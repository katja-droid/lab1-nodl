```python
#przedstawi´c sposoby la˛czenia ramek danych za pomoca˛ metod merge i concat

df_big = pd.DataFrame({
    'age': ['Under 16', '16 to 25 years old', '26 to 35 years old', '36 to 45 years old', 'Over 45 years old'],
    'gp_pre_income': [50000, 20000, 35000, 45000, 60000]
})

df = pd.DataFrame({
    'age': ['Under 16', '16 to 25 years old', '26 to 35 years old', '36 to 45 years old', 'Over 45 years old'],
    'gender': ['Female', 'Male', 'Female', 'Male', 'Female']
})

df_dropped = df.drop(['age'], axis=1)

df_merged = pd.merge(df_big, df, on='age', how='inner')
df_concat = pd.concat([df_big, df], axis=0)
df_concat_cols = pd.concat([df_big, df], axis=1)

print("Pierwsza ramka danych (df_big):")
print(df_big)
print("\nDruga ramka danych (df):")
print(df)
print("\nUsunięto 'age' z df:")
print(df_dropped)
print("\nPołączona ramka danych (df_big i df) na podstawie 'age':")
print(df_merged)
print("\nZłączona ramka danych (df_big i df) pionowo:")
print(df_concat)
print("\nZłączona ramka danych (df_big i df) poziomo:")
print(df_concat_cols)


```

```python
#pokaza´c dodawanie nowych kolumn za pomoca˛ operacji matematycznych
df = pd.DataFrame({
    'age': ['Under 16', '16 to 25 years old', '26 to 35 years old', '36 to 45 years old', 'Over 45 years old'],
    'gp_pre_income': [50000, 20000, 35000, 45000, 60000]
})

# Dodanie nowej kolumny: całkowity dochód (np. z bonusami)
df['total_income'] = df['gp_pre_income'] + 5000  # Dodajemy 5000 jako bonus

print("Ramka danych z nową kolumną (total_income):")
print(df)
```

```python
#przedstawi´c na przyk ladzie dodawanie nowych kolumn z pomoca˛ funkcji lambda
df['age_in_years'] = df['age'].apply(lambda x: 20 if '16 to 25 years old' in x else 
                                      (30 if '26 to 35 years old' in x else 
                                       (40 if '36 to 45 years old' in x else 50)))

print("\nRamka danych z kolumną age_in_years (wyliczenie wieku na podstawie przedziału):")
print(df)
```
